```python
#1. Import packages

from ultralytics import YOLO
import ultralytics
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#2. Load the pretrained model
model = YOLO('yolo11x.pt')
```

```python
#3. Plot out the result for testing the pre-trained model
result2 = model.predict(r'C:\Users\User\Desktop\AI_ML_TRAINING\YPAI09\Capstone\Capstone5(Final)\Dataset\test4.jpg') 
img= result2[0].plot()  
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.xticks([]) 
plt.yticks([])
plt.grid(False)
plt.axis('off')
plt.show()
```

```python
#4. Print Model Architecture
print(model)
```

```python
#5. Training with custom dataset
results= model.train(data=r'C:\Users\User\Desktop\AI_ML_TRAINING\YPAI09\Capstone\Capstone5(Final)\Dataset\vegetables.v8i.yolov11\data.yaml', 
                     epochs=30, 
                     imgsz=240, 
                     augment=True, 
                     shear=0.0, 
                     flipud=0.0, 
                     fliplr=0.5,
                     )

```
